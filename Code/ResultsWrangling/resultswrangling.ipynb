```python
import pickle 
import sys
import os
import pandas as pd
```

```python
# open pickled files in results and append to dictionary
results = []
filenames = []
for root, dirs, files in os.walk("/Users/teissherman/Desktop/GitHub/IndividualProject/Project/Output/MLResults/"):
    for file in files:
        # remove .pickle
        filenames.append(str(file).replace(".pickle", ""))
        with open( root + file, 'rb') as f:
            loadedresults = pickle.load(f)

        #use pathname as lookup in dictionary
        results.append(loadedresults)
```

```python
# new dictionary with all model results
all_results = {}
for index, name in enumerate(filenames):
    all_results[name] = results[index]
```

```python
# convert to dataframe
df_results = pd.DataFrame.from_dict(all_results)
df_results = df_results.T.reset_index()
```

```python
# sort according to highest auc scores
df_melted = pd.melt(df_results, id_vars=['index'], value_vars=['auc_mlp', 'auc_lr', 'auc_rf']).sort_values('value', ascending=False)
df_melted.reset_index(inplace=True)
df_melted.drop(['level_0'], axis=1, inplace=True)
df_melted.rename({'index': 'Configuration'}, axis=1, inplace=True)
```

```python
pd.set_option("display.max_rows", None)
df_melted
```

```python
# pickle all results
with open('/Users/teissherman/Desktop/GitHub/IndividualProject/Project/Output/CosineExperiment/all_auc_scores.pickle', 'wb') as f:
    pickle.dump(df_melted, f)
```

```python
# visualize roc_curves of top 10 auc models
import matplotlib.pyplot as plt
```

```python
# inverse projection rules for initial 40 settings
df_melted_R_plus = df_melted[df_melted['Configuration'].str.contains('R+', regex=False)]
df_melted_R_plus = df_melted_R_plus[df_melted_R_plus['Configuration'].str.contains('100', regex=False)]
df_melted_R_plus.head(10)
```

```python
indices, highest_score_types, model_types = [], [], []
for x in df_melted_R_plus[:5].iterrows():
    index, highest_score_type = x[1][0], x[1][1]
    model_type = highest_score_type.split('_')[1]
    indices.append(index)
    highest_score_types.append(highest_score_type)
    model_types.append(model_type)
```

```python
# plot for top 5 models across all settings
plt.figure()
for index, score, model in zip(indices,highest_score_types,model_types):
    plt.plot(all_results[index]['fpr_' + str(model)], all_results[index]['tpr_' + str(model)], 
    label = index + ' auc_' + str(model) + ':  ' + str(round(all_results[index][score],3)))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Top 5 Performing Document Settings')
plt.legend()
plt.show()
```

```python
plt.figure()
for index, score, model in zip(indices,highest_score_types,model_types):
    plt.plot(all_results[index]['precision_' + str(model)], all_results[index]['recall_' + str(model)], 
    label = index + ' aupr_' + str(model) + ':  ' + str(round(all_results[index]['aupr_' + str(model)],3)))
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.title('Top 5 AUPR Scores')
plt.legend()
plt.show()
```

```python
# mlp results ranked
mlp_results = df_results[['index', 'auc_mlp']].sort_values('auc_mlp', ascending=False)
# rf results ranked
rf_results = df_results[['index', 'auc_rf']].sort_values('auc_rf', ascending=False)
# lr results ranked
lr_results = df_results[['index', 'auc_lr']].sort_values('auc_lr', ascending=False)
```

```python
# sort according to highest aupr scores
df_melted_aupr = pd.melt(df_results, id_vars=['index'], value_vars=['aupr_mlp', 'aupr_lr', 'aupr_rf']).sort_values('value', ascending=False)
```

```python
# f-score calculations across thresholds
precisions = ['precision_mlp','precision_lr','precision_rf']
recalls = ['recall_mlp', 'recall_lr', 'recall_rf']
models = ['mlp', 'lr', 'rf']
for key in all_results.keys():
    for precision, recall, model in zip(precisions,recalls,models):
        precision = all_results[key][precision]
        recall = all_results[key][recall]
        f1 = 2 * ((precision * recall) / (precision + recall))
        all_results[key]['f1_' + str(model)] = f1
```
